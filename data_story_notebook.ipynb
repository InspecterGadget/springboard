```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()


```

```python
def before_after(b_variable):
    if b_variable:
        return 'Before'
    else:
        return 'After'
def before_after1(year):
    if year < 15:
        return 'Before'
    else:
        return 'After'
```

```python
clean_ncaa = pd.read_excel('All Data Clean Rd3.xlsx')
clean_ncaa['SC Change'] =clean_ncaa['Year'].apply(before_after1)
```

```python
fig1_y_list = ['FGA', '3PA', 'AST']
for i in range(3):
    plt.subplot(131+i)
    sns.swarmplot(x = 'SC Change', y = fig1_y_list[i], data = clean_ncaa, size =3)
plt.suptitle("Figure 1: Per Game Stats Before and After the Shot Clock change",
             y=1.02,
             fontsize=14)
plt.rcParams['figure.figsize'] = [10, 5]
plt.tight_layout()
plt.show()
```

```python
plt.subplot(121)
plt.scatter(clean_ncaa['FGA'], clean_ncaa['FG%'], s=20, alpha = .6)
plt.xlabel("Field Goals attempted per game")
plt.ylabel("Percentage of Field Goals made")
plt.subplot(122)
plt.scatter(clean_ncaa['3PA'], clean_ncaa['3P%'],s=20, alpha = .6)
plt.xlabel("3-Point Field Goals attempted per game")
plt.ylabel("Percentage of 3-Point Field Goals made")
plt.suptitle("Figure 2: Per Game Stat Comparisons",
             y=1.02,
             fontsize=14)
plt.rcParams['figure.figsize'] = [15,6]
plt.tight_layout()
plt.show()
```

```python
plt.hist(clean_ncaa['PQ average'], bins = 20)
plt.xlabel("Average Parity")
plt.ylabel("Number of team seasons between 2011-2018")
plt.rcParams['figure.figsize'] = [10, 10]
plt.title("Figure 4: Histogram of NCAA Parity")
plt.show()
```

```python
plt.scatter(clean_ncaa['Win percentage'],clean_ncaa['PQ average'], s=5, alpha = .7)
plt.xlabel('Win Percentage')
plt.ylabel('Average Parity')
plt.rcParams['figure.figsize'] = [8, 8]
plt.title("Figure 5: NCAA Win Percentages against Parity")
plt.show()
```

```python
plt.subplot(131)
plt.scatter(clean_ncaa['FGA'], clean_ncaa['PQ average'], s=20, alpha = .6)
plt.xlabel("Field Goals attempted per game")
plt.ylabel("Parity")
plt.subplot(132)
plt.scatter(clean_ncaa['3PA'], clean_ncaa['PQ average'], s=20, alpha = .6)
plt.xlabel("3-Point Field Goals attempted per game")
plt.ylabel("Parity")
plt.subplot(133)
plt.scatter(clean_ncaa['Tm.'], clean_ncaa['PQ average'], s=20, alpha = .6)
plt.xlabel("Points per game")
plt.ylabel("Parity")
plt.suptitle("Figure 6: Initial Parity Comparisons",
             y=1.02,
             fontsize=14)
plt.rcParams['figure.figsize'] = [28,6]
plt.tight_layout()
plt.show()
```

```python
sns.swarmplot(x= 'SC Change', y= '3PA%', data = clean_ncaa, size = 5)
plt.rcParams['figure.figsize'] = [18, 5]
plt.title("Figure 7: Shooting more 3's Reletive to 2's")
plt.show()
```
